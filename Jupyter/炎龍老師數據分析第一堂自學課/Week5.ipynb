## 用 Groupby找到美國最常目擊UFO的州

```python
df = pd.read_csv("http://bit.ly/uforeports")
```

```python
df.head()
```

## 計算次數
dataframe名稱.count()

## 依項目名稱集合值
dataframe名稱.groupby(項目)

```python
df_state = df.groupby("State").count()
```

```python
df_state
```

```python
df_state.sort_values(by="Time",ascending=False)
```

## sort不產生新的dataframe的方法

dataframe名稱.sort_values(by='要排序的項目', ascending=False, inplace=True)
ascending 升密

```python
df_state.sort_values(by="Time",ascending=False, inplace=True)
```

```python
df_state.head()
```

```python
df_state[:10].Time.plot(kind='bar')
```

```python
import pandas_datareader.data as web
```

```python
df = web.DataReader("AAPL", "yahoo", start ="2013-10-01", end ="2018-10-01")
```

```python
df
```

```python
P = df["Adj Close"]
```

```python
P.head()
```

```python
P.plot()
```

## 投資報酬率

```python
r = P.diff()/P 
```

```python
r[-100:].plot()  # 前一百天
```

## 移動平均
